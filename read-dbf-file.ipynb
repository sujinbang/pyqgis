```python
from dbfread import DBF
import pandas as pd

# DBF 파일 경로
dbf_file_path = "./ETC_도로중심선_1년_1년_2024/csv-to-shp/ETC_S0_07_04_709280.dbf"

try:
    # DBF 파일 열기
    table = DBF(dbf_file_path , encoding='euc-kr')

    # 레코드 출력
    # for record in table:
        # print(record)

    # 또는 데이터프레임으로 변환하여 사용

    df = pd.DataFrame(iter(table))
    print(df)

except FileNotFoundError:
    print(f"Error: File not found at {dbf_file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# expression
for i in range(0, len(df['도로명'].unique())):
    road_name = df['도로명'].unique()[i]
    expression = f'"도로명" = \'{road_name}\''
    print(expression)
```
